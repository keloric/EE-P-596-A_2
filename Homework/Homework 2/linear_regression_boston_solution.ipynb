```python
# TODO
import numpy as np

def least_square(X, y, bias=False):
    if bias:
        X = np.concatenate([X, np.ones((X.shape[0], 1))], axis=1)
    XTy = np.matmul(X.T, y)
    XTX_inv = np.linalg.inv(np.matmul(X.T, X))
    theta = np.matmul(XTX_inv, XTy)
    return theta

def ridge_reg(X, y, eta, bias=False):
    if bias:
        X = np.concatenate([X, np.ones((X.shape[0], 1))], axis=1)
    XTy = np.matmul(X.T, y)
    XTX_inv = np.linalg.inv(np.matmul(X.T, X) + eta*np.eye(X.shape[1]))
    theta = np.matmul(XTX_inv, XTy)
    return theta

# evaluation
def pred_fn(X, theta, bias=False):
    if bias:
        return np.matmul(X, theta[:-1]) + theta[-1]
    else:
        return np.matmul(X, theta)

def root_mean_square_error(pred, y):
    rmse = np.sqrt(np.mean((pred - y)**2))
    return rmse
```

```python
# train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature, price, test_size=0.3, random_state=8)
```

```python
# linear regression
theta = least_square(X_train, y_train)
df_theta = pd.DataFrame(zip(boston.feature_names, theta[:len(boston.feature_names)]),columns=['Feature','Coeff'])
df_theta
```

```python
df_theta_abs = pd.DataFrame(zip(boston.feature_names, np.abs(theta[:len(boston.feature_names)])),columns=['Feature','|Coeff|'])
df_theta_abs.sort_values('|Coeff|', ascending=False)
```

```python
y_train_pred = pred_fn(X_train, theta)
rmse_train = root_mean_square_error(y_train_pred, y_train)

y_test_pred = pred_fn(X_test, theta)
rmse_test = root_mean_square_error(y_test_pred, y_test)

print('RMSE of training set =', rmse_train)
print('RMSE of test set =', rmse_test)
```

```python
y_train_pred_r = pred_fn(X_train, theta_r)
rmse_train_r = root_mean_square_error(y_train_pred_r, y_train)

y_test_pred_r = pred_fn(X_test, theta_r)
rmse_test_r = root_mean_square_error(y_test_pred_r, y_test)

print('RMSE of training set =', rmse_train_r)
print('RMSE of test set =', rmse_test_r)
```

```python
# ridge regression using top-3 features and with normalization
s = [5,10,12]
X_ns = normalize(np.concatenate([X_train[:, s], X_test[:, s]]), norm='l2', axis=0)
X_train_ns = X_ns[:X_train.shape[0]]
X_test_ns = X_ns[X_train.shape[0]:]

theta_ns = ridge_reg(X_train_ns, y_train, 1.0e-3)

y_train_pred_ns = pred_fn(X_train_ns, theta_ns)
rmse_train_ns = root_mean_square_error(y_train_pred_ns, y_train)

y_test_pred_ns = pred_fn(X_test_ns, theta_ns)
rmse_test_ns = root_mean_square_error(y_test_pred_ns, y_test)

print('RMSE of training set =', rmse_train_ns)
print('RMSE of test set =', rmse_test_ns)
```
